```python
# =========================================
# ðŸš€ InstalaÃ§Ã£o de Pacotes NecessÃ¡rios
# =========================================
!pip install -q google-adk google-genai
!pip install pyngrok


# =========================================
# ðŸš€ ImportaÃ§Ãµes
# =========================================
from google.adk.agents import Agent
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.adk.tools import google_search
from google.genai import types
from google.generativeai import client
from datetime import date
import warnings
warnings.filterwarnings("ignore")

# Configura a API Key do Google Gemini

import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

# Configura o cliente da SDK do Gemini

from google import genai

client = genai.Client()

MODEL_ID = "gemini-2.0-flash"

# =========================================
# ðŸš€ FunÃ§Ãµes Auxiliares
# =========================================
def call_agent(agent: Agent, message_text: str) -> str:
    session_service = InMemorySessionService()
    session = session_service.create_session(app_name=agent.name, user_id="user1", session_id="session1")
    runner = Runner(agent=agent, app_name=agent.name, session_service=session_service)
    content = types.Content(role="user", parts=[types.Part(text=message_text)])

    final_response = ""
    for event in runner.run(user_id="user1", session_id="session1", new_message=content):
        if event.is_final_response():
            for part in event.content.parts:
                if part.text is not None:
                    final_response += part.text
                    final_response += "\n"
    return final_response

# =========================================
# ðŸš€ Agente Facilize
# =========================================
def agente_facilize(topico, contexto):
    assistente = Agent(
        name="agente_facilize",
        model=MODEL_ID,
        instruction="""
            VocÃª Ã© a Fa, uma assistente virtual amigÃ¡vel e paciente, especializada em ajudar idosos a se sentirem seguros e confiantes ao usar tecnologias digitais.
            Sempre que possÃ­vel, busque informaÃ§Ãµes atualizadas usando a ferramenta de busca do Google (google_search).
        """,
        description="Assistente digital amigÃ¡vel para idosos, focado em inclusÃ£o e suporte",
        tools=[google_search]
    )

    entrada_do_assistente = f"TÃ³pico: {topico}\nContexto: {contexto}"
    resposta = call_agent(assistente, entrada_do_assistente)
    return resposta

# =========================================
# ðŸš€ ExecuÃ§Ã£o do Chatbot Facilize
# =========================================
data_de_hoje = date.today().strftime("%d/%m/%Y")
print("Iniciando a Assistente Facilize\n")

# Pergunta o dispositivo usado
dispositivo = input("Oi! Eu sou a FÃ¡, sua assistente digital. ðŸ˜Š
 Estou aqui para te ajudar hoje! Para comeÃ§armos, vocÃª pode me dizer qual dispositivo vocÃª estÃ¡ usando? (Celular, Tablet ou Computador) ")
print(f"âœ… Dispositivo escolhido: {dispositivo}\n")

# Pergunta o que deseja aprender
objetivo_escolhido = input("Agora me conte, como posso te ajudar hoje? ")

# Envio da solicitaÃ§Ã£o para o agente
print("ðŸ”Ž Buscando as informaÃ§Ãµes mais recentes para te ajudar, aguarde um momento...\n")
contexto = f"O usuÃ¡rio estÃ¡ utilizando um {dispositivo} e deseja aprender sobre '{objetivo_escolhido}'."
resposta = agente_facilize(objetivo_escolhido, contexto)

# Resposta para o usuÃ¡rio
print(f"\nðŸ’¡ Resposta do Facilize:\n{resposta}")

while True:
    print("\nâœ… VocÃª conseguiu realizar a tarefa? (sim/nÃ£o/explicar de novo)")
    sucesso = input().strip().lower()

    if sucesso == "sim":
        print("Que Ã³timo! ðŸ˜Š Fico feliz em ter ajudado. Caso precise de mais alguma coisa, estarei por aqui!")
        break
    elif sucesso == "nÃ£o":
        print("Sem problemas! Quer tentar de novo ou deseja ajuda em algo diferente? (tentar de novo / ajuda diferente)")
        acao = input().strip().lower()

        if acao == "tentar de novo":
            print("Vou explicar novamente, com mais detalhes.\n")
            resposta = agente_facilize(objetivo_escolhido, contexto)
            print(f"\nðŸ’¡ Resposta do Facilize:\n{resposta}")
        else:
            print("Tudo bem! Me conte como posso te ajudar em outra coisa. ðŸ˜Š")
            break
    elif sucesso == "explicar de novo":
        print("Vou explicar novamente, com mais detalhes.\n")
        resposta = agente_facilize(objetivo_escolhido, contexto)
        print(f"\nðŸ’¡ Resposta do Facilize:\n{resposta}")
    else:
        print("Desculpe, nÃ£o entendi sua resposta. Por favor, digite 'sim', 'nÃ£o' ou 'explicar de novo'.")





```
